```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(test)), sep=',')
```

```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='
def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
#     '''queries carto datasets from a given carto account
#     Arguments: 
#     query - string: a valid sql query string
#     format - outlut format  OPTIONAL (default CSV)
#     source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
#     Returns:
#     the return of the sql query AS A STRING
#     NOTES:
#     designed for the carto API, tested only with CSV return format'''
    
    data = urlencode({'format': format, 'q': query}).encode("utf-8")
    response = urlopen(source, data)
    return response.read()
def get_data(data):
    return pd.read_csv(BytesIO(data), sep=',')
```

# Task 1 — Familiarize with SQL Clauses

### 1.1 Sort data by start_station_id, tripduration
- Only checking trips with duration <= 3 hours

```python
query = '''
SELECT * FROM fb55.citibike 
WHERE tripduration <= 10800
ORDER BY start_station_id ASC, tripduration ASC;
'''
table = get_data(queryCartoDB(query))
table.head(11)
```

## 1.2 Only show the top/last 10 records (aka head and tail in SQL) 

```python
query = '''
SELECT * FROM fb55.citibike 
ORDER BY start_station_id DESC
'''
table = get_data(queryCartoDB(query))
table.head(5)
```

## 1.3 List all unique start_station_id values 

```python
query = '''
SELECT DISTINCT start_station_id FROM fb55.citibike;
'''
table = get_data(queryCartoDB(query))
table.head(5)
```

## 1.4 Count the number of trips, Find the average/min/max trip duration

```python
query = '''
SELECT COUNT(tripduration) as trip_amount,
MAX(tripduration) as max_trip,
MIN(tripduration) as min_trip,
AVG(tripduration) as avg_trip
FROM fb55.citibike;
'''
table = get_data(queryCartoDB(query))
table
```

# Task 2 — Working with date/time

## 2.1 Selecting trips started on Feb-02-2015 only 

```python
query = '''SELECT * FROM fb55.citibike
WHERE EXTRACT(YEAR FROM starttime::date) = 2015
AND
EXTRACT(MONTH FROM starttime::date) = 02
AND
EXTRACT(DAY FROM starttime::date) = 02;
'''
table = get_data(queryCartoDB(query))
table.head(5)
```

## 2.2 Selecting trips started on the weekends
- What are average trip duration during weekends? 

```python
query = '''SELECT AVG(tripduration) as average_trip_duration_weekends
FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime) = 0
OR EXTRACT(DOW FROM starttime) = 1;
'''
table = get_data(queryCartoDB(query))
table.head(5)
```

## 2.3 Can we do the same for weekday 

```python
query = '''SELECT AVG(tripduration) as average_trip_duration_weekdays
FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime) <> 0
AND EXTRACT(DOW FROM starttime) <> 1;
'''
table = get_data(queryCartoDB(query))
table.head(5)
```

# Task 3 — Working with Space

## 3.1 Showing the list of start station locations
- Using GROUP BY 

```python
query = '''SELECT start_station_longitude, start_station_latitude, start_station_id 
FROM fb55.citibike 
GROUP BY start_station_longitude, start_station_latitude, start_station_id;'''
table = get_data(queryCartoDB(query))
table.head(5)
```

## 3.2 Showing the number of trips started per station but only for stations within 500m of Time Square! 

```python
query = '''SELECT
    start_station_longitude,
    start_station_latitude,
    start_station_id,
    COUNT(tripduration)
    FROM fb55.citibike
    WHERE (6371 * acos( cos( radians(40.7577) ) * cos( radians( start_station_latitude ) ) 
    * cos( radians( start_station_longitude ) - radians(-73.9857) ) 
    + sin( radians(40.7577) ) * sin( radians( start_station_latitude ) ) ) ) < 0.5
    GROUP BY start_station_id,start_station_longitude,start_station_latitude;
    '''
table = get_data(queryCartoDB(query))
table
```

# Task 4 — Putting it all together

# 4.1 Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare! 

```python
query = '''SELECT
    start_station_longitude,
    start_station_latitude,
    start_station_id,
    start_station_name,
    AVG(tripduration)
    FROM fb55.citibike
    WHERE (6371 * acos( cos( radians(40.7577) ) * cos( radians( start_station_latitude ) ) 
    * cos( radians( start_station_longitude ) - radians(-73.9857) ) 
    + sin( radians(40.7577) ) * sin( radians( start_station_latitude ) ) ) ) < 0.5
    AND  EXTRACT(DOW FROM starttime) in (0,6)
    GROUP BY start_station_id,start_station_longitude,start_station_latitude,start_station_name
    ORDER BY AVG(tripduration) DESC;
    '''
table = get_data(queryCartoDB(query))
table.head(5)
```

Thus, Broadway & W 49 St station has the longest average trip duration during weekends and within 500m of TimeSquare
